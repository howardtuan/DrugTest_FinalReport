```python
df.drop(['group','Ensembl_transcript','Ensembl_protein','gene_symbol','go_term','RefSeq_ID','Gene_ID'],axis=1,inplace=True)
df.head()
```

```python
df.corr()
```

```python
# Create a heatmap for the correlation matrix
# Calculating correlation matrix
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', cbar=True)
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```python
# Visualizing each column's data
for column in df.columns:
    if df[column].dtype in ['int64', 'float64']:
        plt.figure(figsize=(8, 4))
        sns.histplot(df[column], kde=True, bins=30)
        plt.title(f'Distribution of {column}')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.show()
```

可以從相關係數看出來,在服藥之前基因的表現量,和服藥之後基因的表現量 之間是高度正相關的。而根據log2(ratio)可以看出來od1在整體來說,相對上調比較多,所以 我們認為od1是治療卵巢癌的較佳藥物。然後我們在找出服用了od1藥物之後, 基因表現量前幾名的基因,研究這些基因有哪些特徵。

```python
# 解決 SettingWithCopyWarning 問題，對篩選後的 DataFrame 使用 .loc 方法
filtered_df = pd.read_excel("Gene microarray/ND_4 hr/1H210011803_Normalized_Data.xlsx", sheet_name=4)

# 計算 OD1 與其他三組的正差距總和
filtered_df['Total_Positive_Diff'] = (
    (filtered_df["N-OD1"] - filtered_df["N-ND1"]) +
    (filtered_df["N-OD1"] - filtered_df["N-ND2"]) +
    (filtered_df["N-OD1"] - filtered_df["N-Con"])
)

# 根據 Total_Positive_Diff 排序，選出前 5 個
top_10_high_od1 = filtered_df.nlargest(5, "Total_Positive_Diff")
# 顯示結果
top_10_high_od1
```
