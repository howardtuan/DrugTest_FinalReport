```python
# 解決 SettingWithCopyWarning 問題，對篩選後的 DataFrame 使用 .loc 方法
filtered_df = pd.read_excel("Gene microarray/ND_4 hr/1H210011803_Normalized_Data.xlsx", sheet_name=4)

# 計算 OD1 與其他三組的正差距總和
filtered_df['Total_Positive_Diff'] = (
    (filtered_df["N-OD1"] - filtered_df["N-ND1"]) +
    (filtered_df["N-OD1"] - filtered_df["N-ND2"]) +
    (filtered_df["N-OD1"] - filtered_df["N-Con"])
)

# 根據 Total_Positive_Diff 排序，選出前 5 個
top_10_high_od1 = filtered_df.nlargest(5, "Total_Positive_Diff")
# 顯示結果
top_10_high_od1
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# 讀取數據檔案的最後一個活頁簿
data_last_sheet = pd.read_excel('Gene microarray/ND_4 hr/1H210011803_Normalized_Data.xlsx', sheet_name=-1)

# 選取相關欄位進行分析
columns_of_interest = ['N-OD1', 'N-Con', 'N-ND1', 'N-ND2']
data_filtered = data_last_sheet[columns_of_interest]

# 執行分群
kmeans = KMeans(n_clusters=3, random_state=42)
data_filtered['Cluster'] = kmeans.fit_predict(data_filtered)

# 計算每群的平均表現量與敏感程度
cluster_features = data_filtered.groupby('Cluster').mean()
cluster_features['OD1 Sensitivity'] = cluster_features['N-OD1'] - cluster_features['N-Con']
cluster_features['ND1 Sensitivity'] = cluster_features['N-ND1'] - cluster_features['N-Con']
cluster_features['ND2 Sensitivity'] = cluster_features['N-ND2'] - cluster_features['N-Con']

# 確定每群對哪種藥物最敏感
cluster_features['Most Sensitive Drug (vs CON)'] = cluster_features[
    ['OD1 Sensitivity', 'ND1 Sensitivity', 'ND2 Sensitivity']
].idxmax(axis=1)

# 動態更改縱軸並繪製圖表
sensitivity_labels = ['OD1 Sensitivity', 'ND1 Sensitivity', 'ND2 Sensitivity']
y_axis_mapping = {'OD1 Sensitivity': 'N-OD1', 'ND1 Sensitivity': 'N-ND1', 'ND2 Sensitivity': 'N-ND2'}
colors = ['red', 'blue', 'green']

for sensitivity in sensitivity_labels:
    plt.figure(figsize=(10, 8))
    for cluster in range(3):
        cluster_data = data_filtered[data_filtered['Cluster'] == cluster]
        plt.scatter(
            cluster_data['N-Con'], cluster_data[y_axis_mapping[sensitivity]], 
            label=f"Cluster {cluster}: {sensitivity} = {round(cluster_features.loc[cluster, sensitivity], 2)}",
            color=colors[cluster], alpha=0.6
        )

    plt.title(f'Cluster Analysis with {sensitivity}')
    plt.xlabel('N-Con (Control Group Expression)')
    plt.ylabel(f"{y_axis_mapping[sensitivity]} Expression")
    plt.legend()
    plt.grid(True)
    plt.show()
```
